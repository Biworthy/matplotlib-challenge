## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
MouseStudy_df = pd.merge(mouse_metadata, study_results, how="left", on=["Mouse ID", "Mouse ID"])
# Display the data table for preview
#mouse_metadata
#study_results
MouseStudy_df
```

```python
# Checking the number of mice.
MouseStudy_df["Mouse ID"].nunique()
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
MouseStudy_df.loc[MouseStudy_df.duplicated(subset=['Mouse ID', 'Timepoint']),'Mouse ID'].unique()
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
MouseStudy_df[MouseStudy_df["Mouse ID"]=='g989']
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
CleanMouseStudy = MouseStudy_df.drop(MouseStudy_df[MouseStudy_df["Mouse ID"]=='g989'].index)
CleanMouseStudy
```

```python
# Checking the number of mice in the clean DataFrame.
CleanMouseStudy["Mouse ID"].nunique()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
cms_groups = CleanMouseStudy.groupby('Drug Regimen')


# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
tumor_mean = cms_groups['Tumor Volume (mm3)'].mean()
tumor_median = cms_groups['Tumor Volume (mm3)'].median()
tumor_variance = cms_groups['Tumor Volume (mm3)'].var()
tumor_stdV = cms_groups['Tumor Volume (mm3)'].std()
tumor_sem = cms_groups['Tumor Volume (mm3)'].sem()


# Assemble the resulting series into a single summary dataframe.
statistics_df = pd.DataFrame({
    'Mean': tumor_mean,
    'Median': tumor_median,
    'Variance': tumor_variance,
    'Standard Deviation': tumor_stdV,
    'SEM': tumor_sem
})
statistics_df
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line
statistics_df2 = CleanMouseStudy.groupby("Drug Regimen").agg({'Tumor Volume (mm3)':['mean', 'median','var','std', 'sem']})
statistics_df2
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
grt_tp = CleanMouseStudy.groupby('Mouse ID')['Timepoint'].max()
grt_tp

grt_tp = grt_tp.reset_index()

# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
grt_tp = grt_tp.merge(CleanMouseStudy, how="left", on=["Mouse ID", "Timepoint"])
grt_tp
```

```python
# Put treatments into a list for for loop (and later for plot labels)
regimens = ['Capomulin', 'Ramicane', 'Ceftamin', 'Infubinol']

# Create empty list to fill with tumor vol data (for plotting)
final_tumor_volume = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for drug in regimens:
    
    # Locate the rows which contain mice on each drug and get the tumor volumes
    tumor_volume = grt_tp.loc[grt_tp['Drug Regimen']==drug, 'Tumor Volume (mm3)']
    
    # add subset 
    quaurtiles = tumor_volume.quantile([.25,.5,.75])
    lowerq = quaurtiles[.25]
    upperq = quaurtiles[.75] 
    iqr = upperq - lowerq
    final_tumor_volume.append(tumor_volume)

    
    # Determine outliers using upper and lower bounds
    LowerBound = lowerq - (1.5* iqr)
    UpperBound = upperq + (1.5*iqr)

    
print(f" The lower quartile is {lowerq}")    
print(f" The upper quartile is {upperq}")
print(f" The interquartile range is {iqr}")
print(f" The Values beilow {LowerBound} could be outliers")
print(f" The Values above {UpperBound} could be outliers")
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
plt.boxplot(final_tumor_volume, 
           labels = regimens)

plt.title("Range of Tumor Volume for Drug Regimen")
plt.ylabel("Tumor Volume")
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin

capomulin_table = CleanMouseStudy.loc[CleanMouseStudy["Drug Regimen"] == 'Capomulin']
mouse = capomulin_table[capomulin_table["Mouse ID"] == 's185']
mouse

plt.plot(mouse['Timepoint'], mouse['Tumor Volume (mm3)'])
plt.title("Tumor Volume Over Time")
plt.xlabel("Timepoint")
plt.ylabel("Tumor Volume")
plt.show()

```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
CapAvgStats = capomulin_table.groupby(["Mouse ID"]).mean()
CapAvgStats
plt.title("Average Tumor Volume and Weight")
plt.xlabel("Average Weight")
plt.ylabel("AverageTumor Volume")
plt.scatter(CapAvgStats["Weight (g)"], CapAvgStats["Tumor Volume (mm3)"], marker = "o", facecolors = "red", edgecolors = "black", alpha = .75)
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
corr = round(st.pearsonr(CapAvgStats["Weight (g)"], CapAvgStats["Tumor Volume (mm3)"])[0],2)
model = st.linregress(CapAvgStats["Weight (g)"], CapAvgStats["Tumor Volume (mm3)"])
y_values = CapAvgStats["Weight (g)"] * model[0] + model[1]
plt.scatter(CapAvgStats["Weight (g)"], CapAvgStats["Tumor Volume (mm3)"], edgecolors = "red")
# plot the regression line
plt.title("Average Tumor Volume and Weight with Linear Regression ")
plt.xlabel("Average Weight")
plt.ylabel("AverageTumor Volume")
plt.plot(CapAvgStats["Weight (g)"], y_values, color= "black", alpha=.5)

plt.show()
```
